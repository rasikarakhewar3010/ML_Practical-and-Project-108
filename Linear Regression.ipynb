```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,4,5,4,5])
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
print("Slope (m)", model.coef_[0])
print("Intercept (c)", model.intercept_)
```

```python
y_pred = model.predict(x)
print("Predicted Values: " , y_pred)
```

```python
plt.scatter(x,y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color= 'red',label='Regression Line')
plt.xlabel('Study Houres')
plt.ylabel('Scores')
plt.title('Liniear Regression exalple')
plt.legend()
plt.show()
```

```python
new_hours = np.array([[6]])
predicted_score = model.predict(new_hours)
print(f"Predicted score for studying {new_hours[0][0]} hours: {predicted_score[0]:.2f}")
```

```python
# new example but now lets not use external lib
# Years of Experience (X) and Salary (y)
X = [1, 2, 3, 4, 5]
y = [30, 35, 40, 45, 50]
```

```python
n = len(X)
avg_x = sum(X) /n
avg_y = sum(y) /n
```

```python
num = 0
den = 0 

for i in range(n):
    num +=( X[i] - avg_x)*(y[i]- avg_y)
    den +=(X[i] - avg_x) ** 2
```

```python
m = num/den
c = avg_y - m * avg_x
```

```python
print('Slope (m): ',m)
print('Intercept (c): ',c)
```

```python
y_pred = []

for i in range(n):
    y_pred.append(m* X[i] +c)

print("Predicted salaries:", y_pred)
```

```python
plt.scatter(X, y, color='blue', label='Actual Salaries')
plt.plot(X, y_pred, color='red', label='Best Fit Line')

plt.xlabel('Years of Experience')
plt.ylabel('Salary (in thousands)')
plt.title('Linear Regression: Experience vs Salary')
plt.legend()
plt.show()
```
