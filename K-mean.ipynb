```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate sample data with 4 clusters
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Plot the generated data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Generated Data for K-Means Clustering1")
plt.show()
1
```

```python
# Create KMeans model with K=4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit the model to our data
kmeans.fit(X)
```

```python
# Predicted cluster for each point
y_kmeans = kmeans.predict(X)

# Coordinates of cluster centers
centers = kmeans.cluster_centers_
print("Cluster Centers:\n", centers)
```

```python
plt.figure(figsize=(8,6))

# Plot each cluster with a color
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Plot centroids as red X
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X')

plt.title("K-Means Clustering with K=4")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertia.append(km.inertia_)

plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method to Determine Optimal K')
plt.show()
```
